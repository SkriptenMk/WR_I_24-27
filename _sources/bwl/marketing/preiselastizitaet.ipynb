# Konditionenpolitik

Das zweite der 4 Ps steht für "Price" (Preis). Anstelle von Preispolitik
spricht man allerdings besser von Konditionenpolitik. Diese umfasst
nicht nur den Preis selbst, sondern auch die Zahlungsbedingungen,
Lieferbedingungen und weitere vertragliche Vereinbarungen.

Trotzdem soll hier in erster Linie auf den Spielraum bei der
Preisgestaltung eingegangen werden.  
Der Preis für ein Produkt muss die Kosten decken und einen Gewinn
erzielen, aber er muss auch wettbewerbsfähig sein und die Zahlungsbereitschaft
der Kunden berücksichtigen. Kosten decken bedeutet, dass kurzfristig
mindestens die variablen Kosten gedeckt sein müssen. Langfristig müssen
auch die Fixkosten berücksichtigt werden, um das Unternehmen
nachhaltig zu betreiben. Wettbewerbsfähig ist der Preis, wenn er im
Vergleich zu ähnlichen Produkten auf dem Markt attraktiv (d.h. nicht
deutlich höher als die Preise vergleichbarer Alternativen) ist und die
Zahlungsbereitschaft der Kunden anspricht.

Wie der durch die so beschriebene obere und untere Preisgrenze definierte
Preisspielraum genutzt werden kann, ist nicht zu Letzt davon abhängig,
wie stark die Nachfrage nach dem Produkt auf Preisänderungen reagiert.  
Um diese Reaktion zu analysieren, wird auf ein Instrument aus der
Volkswirtschaftslehre, die Nachfragekurve, zurückgegriffen. Diese zeigt
die Beziehung zwischen dem Preis eines Produkts und der Menge, die
von den Konsumenten nachgefragt wird. In der Regel gilt: Je höher der
Preis, desto geringer die nachgefragte Menge und umgekehrt.  
Graphisch wird das in einem Diagramm mit einer Achse für den Preis und
einer Achse für die nachgefragte Menge dargestellt. Die Nachfragekurve
verläuft in der Regel von links oben nach rechts unten, was die
negative Beziehung zwischen Preis und nachgefragter Menge verdeutlicht.

<div class="alert alert-block alert-warning">
<b>Achtung:</b> Aus den Naturwissenschaften ist man sich gewohnt, dass die Unabhängige Variable auf der x-Achse und die Abhängige Variable auf der y-Achse dargestellt wird. In einem Preis-Mengen-Diagramm ist das aber umgekehrt. Der Preis ist die unabhängige Variable und wird auf der y-Achse abgetragen, während die nachgefragte Menge, welche abhängig vom Preis ist, auf der x-Achse dargestellt wird.
</div>

Um in einem Jupyter Notebook ein Preis-Mengen-Diagramm darzustellen,
werden hier im Beispiel die Bibliotheken 
<a href="https://matplotlib.org/stable/index.html" target="_blank">
Matplotlib
</a>
und 
<a href="https://pandas.pydata.org/" target="_blank">
pandas
</a>
verwendet.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

In einer ersten Darstellung werden die erforderlichen Daten für das
Preis-Mengen-Diagramm in Python-Listen zur Verfügung gestellt.

```python
preise = [p for p in range(11)]
mengen = [(p * -1) + 10 for p in preise]
```

Die beiden Listen `preise` und `mengen` werden als nächstes mit
Matplotlib visualisiert.

```python
plt.figure(
    figsize=(6, 6)
)
plt.plot(mengen, preise, label="Nachfragekurve")
plt.title("Preis-Mengen-Diagramm")
plt.xlabel("Menge")
plt.ylabel("Preis")
plt.legend()
plt.show()
```

Was geschieht aber, wenn es sich um ein Produkt handelt, bei dem die
Nachfrage nicht so stark auf Preisänderungen reagiert? Ein solches
Produkt könnte ein Medikament sein.

```python
medikamente = [(p * -0.75 + 10) for p in preise]
```

```python
plt.figure(
    figsize=(6, 6)
)
plt.plot(medikamente, preise, label="Medikamente")
plt.title("Preis-Mengen-Diagramm")
plt.xlabel("Menge")
plt.ylabel("Preis")
plt.legend()
plt.show()
```

Eine Darstellung, in der nur die Reaktion auf Preisänderungen der
Nachfrage berücksichtigt wird, kann leicht zu Fehlschlüssen führen. Bei
obiger Grafik muss man sehr genau hinschauen, damit man feststellt, dass
die x-Achse erst bei 2 beginnt...

Aus diesem Grund ist es besser die Standardnachfragekurve und die
Medikamentenkurve in der gleichen Grafik darzustellen.

```python
plt.figure(
    figsize=(6, 6)
)
plt.plot(mengen, preise, label="Nachfragekurve")
plt.plot(medikamente, preise, label="Medikamente")
plt.title("Preis-Mengen-Diagramm")
plt.xlabel("Menge")
plt.ylabel("Preis")
plt.legend()
plt.show()
```

Wenn die beiden Kurven in der gleichen Grafik dargestellt werden, sieht
man, dass die Nachfrage nach Medikamenten weniger stark auf
Preisänderungen reagiert als die Standardnachfrage. Dieses Phänomen wird
als Preiselastizität der Nachfrage bezeichnet. Die Preiselastizität der
Nachfrage ($\eta_{Q,P}$) misst, wie stark die nachgefragte Menge ($Q$)
auf Preisänderungen ($P$) reagiert.     
Aber Achtung, die Preiselastizität der Nachfrage ist nicht das gleiche
wie die Steigung der Nachfragekurve.

Die Preiselastizität der Nachfrage wird mit der folgenden Formel
berechnet:

$$
\eta_{Q,P} = \frac{\%\Delta Q}{\%\Delta P} = \frac{
    \frac{(Q_2 - Q_1)}{\frac{(Q_2 + Q_1)}{ 2}}
    }{
    \frac{(P_2 - P_1)}{\frac{(P_2 + P_1)}{ 2}}}
$$

Um zu zeigen, dass die Preiselastizität der Nachfrage nicht das gleiche
ist, wie die Steigung der Nachfragekurve, wird im folgenden die
Preiselastizität der Nachfrage entlang der Nachfragekurve betrachtet.

Dazu werden zuerst die Daten als DataFrame bereitgestellt.

```python
df = pd.DataFrame({
    "Preis": preise,
    "Menge": mengen,
    "Medis": medikamente,
})
df
```

Die Preiselastizität soll für drei Preisniveaus berechnet werden:

$P_1 = 8$, $P_2 = 5$, $P_3 = 2$

In allen drei Fällen soll der Preis um eine Einheit gesenkt werden.

```python
def pe(q1, q2, p1, p2):
    return (((q2 - q1) / ((q2 + q1) / 2)) / 
            ((p2 - p1) / ((p2 + p1) / 2)))
```

```python
pe1 = pe(2, 3, 8, 7)
pe2 = pe(5, 6, 5, 4)
pe3 = pe(8, 9, 2, 1)
```

```python
print(f"Preiselastizität bei P=8: {pe1:.2f}"
      f"\nPreiselastizität bei P=5: {pe2:.2f}"
      f"\nPreiselastizität bei P=2: {pe3:.2f}")
```

Offensichtlich verändert sich die Preiselastizität der Nachfrage mit dem
Preisniveau. Bei höheren Preisen ist die Nachfrage elastischer, während
sie bei niedrigeren Preisen unelastischer wird. 

Grafisch dargestellt sehen die obigen Resultate folgendermassen aus:


```python
p1en = [7, 4, 1]
etas = [-pe1, -pe2, -pe3]
```

```python
fig, ax = plt.subplots()
ax.bar(p1en, etas, width=1, align="edge", color='skyblue')
ax.set_xlabel('Preisniveau')
ax.set_ylabel('Preiselastizität der Nachfrage')
ax.set_title('Preiselastizität der Nachfrage bei verschiedenen Preisniveaus')
plt.show()
```
