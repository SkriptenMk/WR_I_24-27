```python
preise = [p for p in range(11)]
```

```python
preise
```

```python
mengen = [(p * -1 + 10) for p in preise]
mengen
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(
    figsize=(8, 8)
)
plt.plot(mengen, preise, label='Standardnachfrage')
plt.legend()
plt.ylabel('Preis')
plt.xlabel('Menge')
plt.title('Nachfragekurven')
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Preis': preise,
    'Menge': mengen,
}).set_index('Preis')
```

```python
df
```

```python
df['Medis'] = df.index * -.75 + 10
```

```python
df
```

```python
plt.figure(
    figsize=(8, 8)
)

plt.plot(df['Menge'], df.index, label='Standardnachfrage')
plt.plot(df['Medis'], df.index, label='Medikamentennachfrage')

plt.legend()
plt.ylabel('Preis')
plt.xlabel('Menge')
```
