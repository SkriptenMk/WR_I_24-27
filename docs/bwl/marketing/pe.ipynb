```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
preise = [i for i in range(11)]
```

```python
mengen = [(p * -1 + 10) for p in preise]
mue = [(p * -0.75 + 10) for p in preise]
me = [(p * -1.25 + 10) for p in preise]
```

```python
df = pd.DataFrame({
    'Preis': preise,
    'Menge': mengen,
    'unelastisch' : mue,
    'elastisch' : me,
}).set_index('Preis')
```

```python
df
```

```python
df_filtered = df[df['unelastisch'] >= 0]
```

```python
plt.figure(figsize=(8, 8))

plt.plot(df_filtered['Menge'], df_filtered.index, label='Standardnachfrage')
plt.plot(df_filtered['unelastisch'], df_filtered.index, label='unelastische Nachfrage')
plt.plot(df_filtered['elastisch'], df_filtered.index, label='elastische Nachfrage')


plt.xlabel('Menge')
plt.ylabel('Preis')
plt.legend()
plt.show()
```

Die Preiselastizitt der Nachfrage ($\eta_{Q,P}$) misst, wie stark die nachgefragte Menge ($Q$) auf Preisnderungen ($P$) reagiert. 


$$
\eta_{Q,P} = \frac{\%\Delta Q}{\%\Delta P} = \frac{
    \frac{(Q_2 - Q_1)}{\frac{(Q_2 + Q_1)}{ 2}}
    }{
    \frac{(P_2 - P_1)}{\frac{(P_2 + P_1)}{ 2}}}
$$
