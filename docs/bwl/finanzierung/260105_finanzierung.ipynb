## Ausgangslage

**Das Unternehmen:**

Die Rigi Bahnen AG blickt auf ein erfolgreiches Geschäftsjahr 2024
zurück. Mit einem Umsatz von rund CHF 37.5 Mio. und einem EBITDA von
CHF 12.3 Mio. steht das Unternehmen finanziell sehr gesund da. Die
Eigenkapitalquote ist auf stolze 61% gestiegen.

**Das Projekt «Seebahn Arth»:**

Im Rahmen des 150-Jahr-Jubiläums der Arth-Rigi-Bahn wird eine visionäre
Idee diskutiert: Die Wiederherstellung des historischen
Geleisanschlusses direkt an den Schiffssteg in Arth am See. Dies würde
das Umsteigen vom Schiff auf die Bahn nahtlos ermöglichen und die
Attraktivität für Touristen massiv steigern. 

**Investitionsbedarf:**

Für Planung, Landerwerb, Geleisbau und Anpassung der Perronanlagen wird
mit einem Investitionsvolumen von CHF 15 Mio. gerechnet.

**Die Herausforderung:**

Die Geschäftsleitung beauftragt Sie als externe Finanzberater, eine
optimale Finanzierungsstruktur für dieses Projekt zu erarbeiten. Dabei
sind die Interessen der Aktionäre, die Sicherheit des Unternehmens und
die langfristige Rentabilität gegeneinander abzuwägen.

## Aufgabenstellung

### Teil A: Analyse der Ist-Situation

1. **Sicherheit:** Berechnen Sie die aktuelle Eigenkapitalquote und den Anlagedeckungsgrad 2.
2. **Entschuldungsfähigkeit:** Berechnen Sie den Verschuldungsfaktor
   (Nettoverschuldung / EBITDA). (Hinweis: Nettoverschuldung = verzinzl.
   FK - Flüssige Mittel). 

### Teil B: Finanzierungsoptionen

Diskutieren Sie drei Möglichkeiten, wie die 15 Mio. CHF beschafft werden könnten:

* Option 1: Eigenfinanzierung (Aktienkapitalerhöhung)
* Option 2: Fremdfinanzierung (Bankdarlehen zu ca. 2.5%)
* Option 3: Innenfinanzierung

### Teil C: Die «optimale» Finanzierung & Rendite

Was passiert mit der Eigenkapitalrendite (ROE), wenn das Projekt über
Fremdkapital finanziert wird, im Vergleich zur Eigenfinanzierung? Die
Gesamtkapitalrendite (ROI) liegt aktuell bei ca. 8.7%.
